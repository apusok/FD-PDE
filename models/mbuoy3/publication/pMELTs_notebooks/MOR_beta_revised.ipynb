## Petrological estimate of beta - coefficient of chemical expansion

by A. Pusok, Nov 5, 2021

Density changes due to: temperature, pressure, composition, mineralogy (through reactions and modal proportion of minerals), volume fraction of melt (porosity).

Let's assume this can be written as the total density change: 

<img src="https://latex.codecogs.com/svg.image?d\rho=\frac{\partial&space;\rho}{\partial&space;x_i}&space;dx_i" title="" />

where xi is [T,P,C,M,phi]. We consider temperature, pressure and composition (expressed by the Mg#) here. We consider 2 simplified cases: 

### 1. Isobaric conditions (P=ct).
The density change is:

<img src="https://latex.codecogs.com/svg.image?d\rho&space;=&space;\left(\frac{\partial\rho}{\partial&space;T}\right)_P&space;dT&space;&plus;&space;\left(\frac{\partial\rho}{\partial&space;C}\right)_P&space;dC" title="d\rho = \left(\frac{\partial\rho}{\partial T}\right)_P dT + \left(\frac{\partial\rho}{\partial C}\right)_P dC" />

where the following coefficients are defined:

<img src="https://latex.codecogs.com/svg.image?\alpha&space;=&space;-\frac{1}{\rho}\left(\frac{\partial\rho}{\partial&space;T}\right)_P&space;" title="\alpha = -\frac{1}{\rho}\left(\frac{\partial\rho}{\partial T}\right)_P " />

<img src="https://latex.codecogs.com/svg.image?\beta_P&space;=&space;-\frac{1}{\rho}\left(\frac{\partial\rho}{\partial&space;C}\right)_P&space;" title="\beta_P = -\frac{1}{\rho}\left(\frac{\partial\rho}{\partial C}\right)_P " />

where alpha is coefficient of thermal expansion and betaP is coefficient of chemical expansion at isobaric conditions.

The density equation of state becomes:

<img src="https://latex.codecogs.com/svg.image?\rho&space;=&space;\rho_0&space;\exp{(-\alpha\Delta&space;T&space;-&space;\beta_P\Delta&space;C)}&space;\approx&space;\rho_0(1-\alpha\Delta&space;T&space;-&space;\beta_P\Delta&space;C)" title="\rho = \rho_0 \exp{(-\alpha\Delta T - \beta_P\Delta C)} \approx \rho_0(1-\alpha\Delta T - \beta_P\Delta C)" />

where we used the approximation e^a~(1+a).

For a small T,C increments: (T1,C1)->(T2,C2):
rho2=rho1(1-alpha(T2-T1)-beta(C2-C1))

### 2. Isothermal conditions (T=ct).
The density change is:

<img src="https://latex.codecogs.com/svg.image?d\rho&space;=&space;\left(\frac{\partial\rho}{\partial&space;P}\right)_T&space;dP&space;&plus;&space;\left(\frac{\partial\rho}{\partial&space;C}\right)_T&space;dC" title="" />

where the following coefficients are defined:

<img src="https://latex.codecogs.com/svg.image?k_T&space;=&space;\frac{1}{\rho}\left(\frac{\partial&space;\rho}{\partial&space;P}\right)_T" title="" />

<img src="https://latex.codecogs.com/svg.image?\beta_T&space;=&space;-\frac{1}{\rho}\left(\frac{\partial\rho}{\partial&space;C}\right)_T&space;" title="" />

where k_T is isothermal compressibility and betaT is coefficient of chemical expansion at isothermal conditions.

The density equation of state becomes:

<img src="https://latex.codecogs.com/svg.image?\rho&space;=&space;\rho_0&space;\exp{(k_T\Delta&space;P&space;-&space;\beta_T\Delta&space;C)}&space;\approx&space;\rho_0(1+k_T\Delta&space;P&space;-&space;\beta_T\Delta&space;C)" title="" />

### 3. Isentropic conditions (S=ct).
The density change is:

<img src="https://latex.codecogs.com/svg.image?d\rho&space;=&space;\left(\frac{\partial\rho}{\partial&space;P}\right)_S&space;dP&space;&plus;&space;\left(\frac{\partial\rho}{\partial&space;T}\right)_S&space;dT&space;&plus;&space;\left(\frac{\partial\rho}{\partial&space;C}\right)_S&space;dC" title="" />

where the following coefficients are defined:

<img src="https://latex.codecogs.com/svg.image?k_S&space;=&space;\frac{1}{\rho}\left(\frac{\partial&space;\rho}{\partial&space;P}\right)_S" title="" />

<img src="https://latex.codecogs.com/svg.image?\alpha_S&space;=&space;-\frac{1}{\rho}\left(\frac{\partial\rho}{\partial&space;T}\right)_S&space;" title="" />

<img src="https://latex.codecogs.com/svg.image?\beta_S&space;=&space;-\frac{1}{\rho}\left(\frac{\partial\rho}{\partial&space;C}\right)_S&space;" title="" />

where k_S is isentropic compressibility, alpha_S isentropic coefficient of thermal expansion (alpha_S~alpha_P) and betaS is coefficient of chemical expansion at isentropic conditions.

The density equation of state becomes:

<img src="https://latex.codecogs.com/svg.image?\rho&space;=&space;\rho_0&space;\exp{(k_S\Delta&space;P&space;-&space;\alpha_S\Delta&space;T-&space;\beta_S\Delta&space;C)}&space;\approx&space;\rho_0(1+k_S\Delta&space;P&space;-&space;\alpha_S\Delta&space;T-&space;\beta_S\Delta&space;C)" title="" />

Notes:
* In calculations below, assumed that kS and kT are the same.
* Assumed that alphaP = alphaS
* Integrating the density equation, the limits are (P0,T0,C0,S0) to (P,T,C,S). Generally, (P0,T0,C0,S0) represent surface conditions at which experiments are done (difficult to obtain), but here we assume (P0,T0,C0,S0) to be conditions from ith-1 (previous) pMELTS iteration.
* Actually, not clear which alpha, k is determined by pMELTS, so this could be a source of error.

## pMELTS calculations

The compositional effect on density only takes place during melting, so we impose a melting degree between 0-30%.

### Import modules and set pMELTS version:

```python
from thermoengine import equilibrate
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import pickle
%matplotlib inline

rc('font',**{'family':'serif','serif':['Times new roman']})
rc('text', usetex=True)

# MELTS version - pMELTS
melts  = equilibrate.MELTSmodel(version="5.6.1")
```

```python
# create empty class for each of the melting curves
class EmptyStruct:
    pass

def create_pmelts_object(bulk_comp):
    melts  = equilibrate.MELTSmodel(version="5.6.1")
    
    if (bulk_comp == 'MM3'):
        melts.set_bulk_composition({'SiO2':45.47, 'TiO2':0.11, 'Al2O3':4.0, 'Fe2O3':0.68,'Cr2O3':0.68, 
                                  'FeO': 6.61, 'MnO':0.0, 'MgO':38.53, 'NiO':0.0,'CoO':0.0,
                                  'CaO': 3.59, 'Na2O':0.31, 'K2O':0.0, 'P2O5':0.0, 'H2O':0.0})
    if (bulk_comp == 'mBas'):
        melts.set_bulk_composition({'SiO2':45.26, 'TiO2':0.19, 'Al2O3':3.88,'Fe2O3':0.40,'Cr2O3':0.41, 
                                   'FeO':7.81, 'MnO':0.0, 'MgO':37.85, 'NiO':0.0, 'CoO':0.0,
                                   'CaO':3.57, 'Na2O':0.30, 'K2O':0.01, 'P2O5':0.0, 'H2O':0.0})
    if (bulk_comp == 'KR4003'):
        melts.set_bulk_composition({'SiO2':44.9, 'TiO2':0.16, 'Al2O3':4.26,'Fe2O3':0.0,'Cr2O3':0.41, 
                                   'FeO':8.02, 'MnO':0.13, 'MgO':37.3, 'NiO':0.24, 'CoO':0.0,
                                   'CaO':3.45, 'Na2O':0.22, 'K2O':0.09, 'P2O5':0.0, 'H2O':0.0}) # requires deep melting
    
    b = melts.get_phase_inclusion_status()
    melts.set_phase_inclusion_status({'Actinolite':False, 'Aegirine':False, \
                                  'Aenigmatite':False, 'Akermanite':False, 'Andalusite':False, \
                                  'Anthophyllite':False, 'Apatite':False, 'Biotite':False, 'Chromite':False, \
                                  'Coesite':False, 'Corundum':False, 'Cristobalite':False, 'Cummingtonite':False, \
                                  'Fayalite':False, 'Forsterite':False, 'Gehlenite':False, 'Hematite':False, \
                                  'Hornblende':False, 'Ilmenite':False, 'Ilmenite ss':False, 'Kalsilite':False, \
                                  'Kalsilite ss':False, 'Kyanite':False, 'Leucite':False, 'Lime':False, \
                                  'Liquid Alloy':False, 'Magnetite':False, 'Melilite':False, 'Muscovite':False, \
                                  'Nepheline':False, 'Nepheline ss':False, 'OrthoOxide':False, 'Panunzite':False, \
                                  'Periclase':False, 'Perovskite':False, 'Phlogopite':False, 'Quartz':False, \
                                  'Rutile':False, 'Sanidine':False, 'Sillimanite':False, 'Solid Alloy':False, \
                                  'Sphene':False, 'Tridymite':False, 'Whitlockite':False})

    # Save phases included:
    Phases_all = []
    a = melts.get_phase_inclusion_status()
    for phase in b.keys():
        if a[phase] == True:
            #print ("{0:<15s} Before: {1:<5s} After: {2:<5s}".format(phase, repr(b[phase]), repr(a[phase])))
            Phases_all.append(phase)

    return melts, Phases_all

def get_wt_frac(melts,root,Phases_solid):
    n = len(Phases_solid)
    x = np.zeros(n) # wt frac
    xs = 0.0
    for i in range (0, n):
        x[i]   = melts.get_property_of_phase(root, Phases_solid[i], property_name='Mass')
        xs = xs + x[i]
    for i in range (0, n):
        x[i]   = x[i]/xs
    return x

def get_solid_density(melts,root,Phases_solid):
    # 1/rho = sum(xi/rhoi)
    n = len(Phases_solid)
    rho = np.zeros(n)
    x = get_wt_frac(melts,root,Phases_solid) # wt frac
    
    rhos_inv = 0.0
    for i in range (0, n):
        rho[i] = melts.get_property_of_phase(root, Phases_solid[i], property_name='Density')
        if (x[i]>0.0):
            rhos_inv= rhos_inv+x[i]/rho[i]
    rhos = 1.0/rhos_inv
    return rhos

def get_solid_alpha(melts,root,Phases_solid):
    # alpha = sum(xi*alphai)
    n = len(Phases_solid)
    x = get_wt_frac(melts,root,Phases_solid) # wt frac
    alpha = np.zeros(n)

    alpha_s = 0
    for i in range (0, n):
        alpha[i] = melts.get_property_of_phase(root, Phases_solid[i], property_name='Alpha')
        if (x[i]>0.0):
            alpha_s= alpha_s+x[i]*alpha[i]
    return alpha_s

def get_liquid_fraction(melts,root):
    phase = 'Liquid'
    phi = melts.get_property_of_phase(root, phase, property_name='Mass')/melts.get_property_of_phase(root, 'System', property_name='Mass')
    return phi

def get_liquid_density(melts,root):
    rho = 0.0
    phase = 'Liquid'
    phi = get_liquid_fraction(melts,root)
    if (phi>0.0):
        rho = melts.get_property_of_phase(root, phase, property_name='Density')
    return rho

def get_liquid_alpha(melts,root):
    alpha = 0.0
    phase = 'Liquid'
    phi = get_liquid_fraction(melts,root)
    if (phi>0.0):
        alpha = melts.get_property_of_phase(root, phase, property_name='Alpha')
    return alpha

def get_liquid_compressibility(melts,root):
    k = 0.0
    phase = 'Liquid'
    phi = get_liquid_fraction(melts,root)
    if (phi>0.0):
        k = melts.get_property_of_phase(root, phase, property_name='Beta') # pmelts returns k in 1/bar
    return k/0.1 # in 1/MPa

def get_solid_compressibility(melts,root,Phases_solid):
    # k = sum(xi*ki)
    n = len(Phases_solid)
    x = get_wt_frac(melts,root,Phases_solid) # wt frac
    k = np.zeros(n)

    k_s = 0
    for i in range (0, n):
        k[i] = melts.get_property_of_phase(root, Phases_solid[i], property_name='Beta') # pmelts returns k in 1/bar
        if (x[i]>0.0):
            k_s= k_s+x[i]*k[i]
    return k_s/0.1 # in 1/MPa

def get_liquid_Mg_number(melts,root):
    phase = 'Liquid'
    M_FeO = 72 # molecular mass
    M_MgO = 40
    Mgf   = 0.0 
    phi = get_liquid_fraction(melts,root)
    if (phi>0.0):
        oxide = melts.get_composition_of_phase(root, phase_name=phase, mode='oxide_wt')
        FeO = oxide["FeO"]
        MgO = oxide["MgO"]
        xFe = FeO/M_FeO
        xMg = MgO/M_MgO
        Mgf = xMg/(xMg+xFe)
    return Mgf

def get_solid_Mg_number(melts,root,Phases_solid):
    # FeO = sum(xi*FeOi) - same for MgO
    n = len(Phases_solid)
    x = get_wt_frac(melts,root,Phases_solid) # wt frac
    M_FeO  = 72 # molecular mass
    M_MgO  = 40
    Mgs    = 0.0 
    
    FeO_s = 0
    MgO_s = 0
    for i in range (0, n):
        oxide = melts.get_composition_of_phase(root, phase_name=Phases_solid[i], mode='oxide_wt')
        if (oxide):
            FeO_s = FeO_s + x[i]*oxide["FeO"]
            MgO_s = MgO_s + x[i]*oxide["MgO"]

    xFe = FeO_s/M_FeO
    xMg = MgO_s/M_MgO
    Mgs = xMg/(xMg+xFe)       
    return Mgs
    
def get_density_Mg_liquid_solid(melts,root,T,P,Phases_all):
    # get solid phases
    Phases_solid = Phases_all.copy()
    Phases_solid.remove('Liquid')
    
    # calculate liquid and solid properties
    phi  = get_liquid_fraction(melts,root)
    rhof = get_liquid_density(melts,root)
    rhos = get_solid_density(melts,root,Phases_solid)
    Mgf  = get_liquid_Mg_number(melts,root)
    Mgs  = get_solid_Mg_number(melts,root,Phases_solid)
    alpha_f = get_liquid_alpha(melts,root)
    alpha_s = get_solid_alpha(melts,root,Phases_solid)
    k_f = get_liquid_compressibility(melts,root)
    k_s = get_solid_compressibility(melts,root,Phases_solid)
    return phi, rhos, rhof, Mgs, Mgf, alpha_s, alpha_f, k_s, k_f

def calculate_betaP(rho0,rho1,alpha,DT,DC):
    if (abs(rho0*DC)>1e-10):
        beta = (rho0*(1.0-alpha*DT)-rho1)/(rho0*DC)
    else:
        beta = 0.0
    return beta

def calculate_betaT(rho0,rho1,k,DP,DC):
    if (abs(rho0*DC)>1e-10):
        beta = (rho0*(1.0+k*DP)-rho1)/(rho0*DC)
    else:
        beta = 0.0
    return beta

def calculate_betaS(rho0,rho1,k,alpha,DP,DT,DC):
    if (abs(rho0*DC)>1e-10):
        beta = (rho0*(1.0+k*DP-alpha*DT)-rho1)/(rho0*DC)
    else:
        beta = 0.0
    return beta
```

### 1. Isobaric conditions (P=ct)

```python
def run_isobaric_melting(bulk_comp,Tstart,Tfinal,Pstart):
    sim=EmptyStruct()
    sim.bulk_comp = bulk_comp
    melts, Phases_all = create_pmelts_object(sim.bulk_comp)
    
    # nsteps = 200
    p_increment = 0.0
    t_increment = 1.0
    nsteps = int((Tfinal-Tstart)/t_increment)

    sim.p   = np.zeros(nsteps)
    sim.t   = np.zeros(nsteps)
    sim.phi = np.zeros(nsteps)
    sim.rhof= np.zeros(nsteps)
    sim.rhos= np.zeros(nsteps)
    sim.Mgf = np.zeros(nsteps)
    sim.Mgs = np.zeros(nsteps)
    sim.alphaf= np.zeros(nsteps)
    sim.alphas= np.zeros(nsteps)
    sim.kf    = np.zeros(nsteps)
    sim.ks    = np.zeros(nsteps)
    sim.betaf = np.zeros(nsteps-1)
    sim.betas = np.zeros(nsteps-1)
    sim.phases= Phases_all

    sim.p[0] = Pstart
    sim.t[0] = Tstart

    for i in range (1,nsteps):
        sim.p[i] = sim.p[i-1]+p_increment
        sim.t[i] = sim.t[i-1]+t_increment

    for i in range (0,nsteps):
        if (i==0):
            output = melts.equilibrate_tp(sim.t[i],sim.p[i], initialize=True)
        else:
            output = melts.equilibrate_tp(sim.t[i],sim.p[i])
        (status, t, p, xmlout) = output[0]
        sim.phi[i], sim.rhos[i], sim.rhof[i], sim.Mgs[i], sim.Mgf[i], sim.alphas[i], sim.alphaf[i], sim.ks[i], sim.kf[i] = get_density_Mg_liquid_solid(melts,xmlout,sim.t[i],sim.p[i],Phases_all)
        if (i>0):
            DT  = sim.t[i]-sim.t[i-1]
            DCF = sim.Mgf[i]-sim.Mgf[i-1]
            DCS = sim.Mgs[i]-sim.Mgs[i-1]
            sim.betaf[i-1] = calculate_betaP(sim.rhof[i-1],sim.rhof[i],sim.alphaf[i],DT,DCF)
            sim.betas[i-1] = calculate_betaP(sim.rhos[i-1],sim.rhos[i],sim.alphas[i],DT,DCS)
    return sim

# simulations - assume 1deg increment
sim0 = run_isobaric_melting('MM3',1240,1325,500) # phi [0.05583931 - 0.30363128]
sim1 = run_isobaric_melting('MM3',1250,1420,1000) # phi [0.00455226 - 0.30010121]
sim2 = run_isobaric_melting('MM3',1350,1490,1500) # phi [0.00375698 - 0.29942145]
sim3 = run_isobaric_melting('MM3',1420,1540,2000) # phi [0.00109972 - 0.2950253]

# save pickle data
fname_pickle = 'sim_1_isobaric.pickle'
with open(fname_pickle, 'wb') as fh:
    pickle.dump(sim0,fh)
    pickle.dump(sim1,fh)
    pickle.dump(sim2,fh)
    pickle.dump(sim3,fh)

# print(sim3.phi)
```

### 2. Isothermal conditions (T=ct)

```python
def run_isothermal_melting(bulk_comp,Pstart,Pfinal,Tstart):
    sim=EmptyStruct()
    sim.bulk_comp = bulk_comp
    melts, Phases_all = create_pmelts_object(sim.bulk_comp)
    
    # nsteps = 200
    p_increment = -10.0
    t_increment = 0.0
    nsteps = int((Pfinal-Pstart)/p_increment)

    sim.p   = np.zeros(nsteps)
    sim.t   = np.zeros(nsteps)
    sim.phi = np.zeros(nsteps)
    sim.rhof= np.zeros(nsteps)
    sim.rhos= np.zeros(nsteps)
    sim.Mgf = np.zeros(nsteps)
    sim.Mgs = np.zeros(nsteps)
    sim.alphaf= np.zeros(nsteps)
    sim.alphas= np.zeros(nsteps)
    sim.kf    = np.zeros(nsteps)
    sim.ks    = np.zeros(nsteps)
    sim.betaf = np.zeros(nsteps-1)
    sim.betas = np.zeros(nsteps-1)
    sim.phases= Phases_all

    sim.p[0] = Pstart
    sim.t[0] = Tstart

    for i in range (1,nsteps):
        sim.p[i] = sim.p[i-1]+p_increment
        sim.t[i] = sim.t[i-1]+t_increment

    for i in range (0,nsteps):
        if (i==0):
            output = melts.equilibrate_tp(sim.t[i],sim.p[i], initialize=True)
        else:
            output = melts.equilibrate_tp(sim.t[i],sim.p[i])
        (status, t, p, xmlout) = output[0]
        sim.phi[i], sim.rhos[i], sim.rhof[i], sim.Mgs[i], sim.Mgf[i], sim.alphas[i], sim.alphaf[i], sim.ks[i], sim.kf[i] = get_density_Mg_liquid_solid(melts,xmlout,sim.t[i],sim.p[i],Phases_all)
        if (i>0):
            DP  = sim.p[i]-sim.p[i-1] # MPa
            DCF = sim.Mgf[i]-sim.Mgf[i-1]
            DCS = sim.Mgs[i]-sim.Mgs[i-1]
            sim.betaf[i-1] = calculate_betaT(sim.rhof[i-1],sim.rhof[i],sim.kf[i],DP,DCF)
            sim.betas[i-1] = calculate_betaT(sim.rhos[i-1],sim.rhos[i],sim.ks[i],DP,DCS)
    return sim

# simulations 
del sim0, sim1, sim2, sim3
sim0 = run_isothermal_melting('MM3',1210,400,1300) # phi [0.00550759 - 0.29815962]
sim1 = run_isothermal_melting('MM3',1500,600,1350) # phi [0.00375698 - 0.3078587]
sim2 = run_isothermal_melting('MM3',1800,870,1400) # phi [0.00438613 - 0.30354621]
sim3 = run_isothermal_melting('MM3',2250,1180,1450) # phi [0.00055071 - 0.30463909]

# save pickle data
fname_pickle = 'sim_2_isothermal.pickle'
with open(fname_pickle, 'wb') as fh:
    pickle.dump(sim0,fh)
    pickle.dump(sim1,fh)
    pickle.dump(sim2,fh)
    pickle.dump(sim3,fh)

# print(sim1.phi)
```

### 3. Isentropic conditions (adiabatic, S=ct)

```python
def run_isentropic_melting(bulk_comp,Tstart,Pstart,Pfinal):
    sim=EmptyStruct()
    sim.bulk_comp = bulk_comp

    # create new pMELTS object with a given bulk composition
    melts, Phases_all = create_pmelts_object(sim.bulk_comp)
    output = melts.equilibrate_tp(Tstart, Pstart, initialize=True)
    (status, t, p, xmlout) = output[0]

    # compute entropy S (J/K) at specified P (MPa)
    Sstart = melts.get_property_of_phase(xmlout,'System', 'Entropy')

    # run the sequence of calculations along a S=constant, P path
    # nsteps = 200
    s_increment = 0.0
    p_increment = -10.0
    nsteps = int((Pfinal-Pstart)/p_increment)

    sim.p   = np.zeros(nsteps)
    sim.s   = np.zeros(nsteps)
    sim.t   = np.zeros(nsteps)
    sim.phi = np.zeros(nsteps)
    sim.rhof= np.zeros(nsteps)
    sim.rhos= np.zeros(nsteps)
    sim.Mgf = np.zeros(nsteps)
    sim.Mgs = np.zeros(nsteps)
    sim.alphaf= np.zeros(nsteps)
    sim.alphas= np.zeros(nsteps)
    sim.kf    = np.zeros(nsteps)
    sim.ks    = np.zeros(nsteps)
    sim.betaf = np.zeros(nsteps-1)
    sim.betas = np.zeros(nsteps-1)

    sim.p[0] = Pstart
    sim.t[0] = Tstart
    sim.s[0] = Sstart

    for i in range (1,nsteps):
        sim.p[i] = sim.p[i-1]+p_increment
        sim.s[i] = sim.s[i-1]+s_increment

    for i in range (0,nsteps):
        if (i==0):
            output = melts.equilibrate_sp(sim.s[i],sim.p[i], initialize=True)
        else:
            output = melts.equilibrate_sp(sim.s[i],sim.p[i])
        (status, t, p, xmlout) = output[0]
        sim.t[i] = t
        sim.phi[i], sim.rhos[i], sim.rhof[i], sim.Mgs[i], sim.Mgf[i], sim.alphas[i], sim.alphaf[i], sim.ks[i], sim.kf[i] = get_density_Mg_liquid_solid(melts,xmlout,sim.t[i],sim.p[i],Phases_all)
    
        if (i>0):
            DP  = sim.p[i]-sim.p[i-1] # MPa
            DT  = sim.t[i]-sim.t[i-1]
            DCF = sim.Mgf[i]-sim.Mgf[i-1]
            DCS = sim.Mgs[i]-sim.Mgs[i-1]
            sim.betaf[i-1] = calculate_betaS(sim.rhof[i-1],sim.rhof[i],sim.kf[i],sim.alphaf[i],DP,DT,DCF)
            sim.betas[i-1] = calculate_betaS(sim.rhos[i-1],sim.rhos[i],sim.ks[i],sim.alphas[i],DP,DT,DCS)
    return sim

# isentropic simulations
# del sim0, sim1, sim2, sim3
del sim0, sim1
sim0 = run_isentropic_melting('MM3',1420,2000,400) # Tstart, Pstart, Pfinal; phi [0.00109972 - 0.29168933]
sim1 = run_isentropic_melting('mBas',1380,2000,300) # Tstart, Pstart, Pfinal; phi [0.00150459 - 0.28108829]
# sim2 = run_isentropic_melting('KR4003',1500,4000,1000) # Tstart, Pstart, Pfinal; phi [] - requires deep melting

# save pickle data
fname_pickle = 'sim_3_isentropic.pickle'
with open(fname_pickle, 'wb') as fh:
    pickle.dump(sim0,fh)
    pickle.dump(sim1,fh)

# print(sim0.phi)
```
